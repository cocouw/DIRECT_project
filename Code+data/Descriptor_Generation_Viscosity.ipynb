```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from __future__ import print_function
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors.MoleculeDescriptors import MolecularDescriptorCalculator as Calculator
```

```python
data = pd.DataFrame.from_csv('Viscosity.csv')
```

```python
data.head()
```

```python
n = data.shape[0]
```

```python
f = open('Deslist','r')
Deslist = []
for line in f:
    Deslist.append(line.strip('\n\t'))
#print(Deslist)
calc = Calculator(Deslist)
D = len(Deslist)
d = len(Deslist)*2+3
print(n,d)
```

```python
X = np.zeros((n,d))
X[:,-3] = data['T']
X[:,-2] = data['P']
for i in range(n):
    cation = Chem.MolFromSmiles(data['SMILE_CAT'][i])
    anion  = Chem.MolFromSmiles(data['SMILE_ANI'][i])
    X[i][:D]    = calc.CalcDescriptors(cation)
    X[i][D:2*D] = calc.CalcDescriptors(anion)
X[:,-1] = data['Viscosity']
np.savetxt("Viscosity.txt",X)
```
